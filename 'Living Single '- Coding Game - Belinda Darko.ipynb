```python
# Game inspired by sitcom TV Show called Living Single S3E16
"""
Created on Fri Aug 22 11:23 AM 2021

@author: belinda.darko
"""

"""
    DocString:
    
    A) Introduction
    This game is based on the sitcome TV Show Living Single
    season 3 episode 16. 
    See link: https://www.imdb.com/title/tt0106056/episodes?season=3
    
    It consists of five stages/scenarios where the
    end goal is to get to the final stage to see what happens
    in the story. The game has defined functions for 
    start, win, and lose. This game requires you to engage in
    this romantic story as we see what unfolds between the characters.
    
    Scenario 1: On set
    Scenario 2: On the elevator
    Scenario 3: Date #1 with first guy
    Scenario 4: Date #2 with second guy
    Scenario 5: At the birthday party
    
    
    B) Known Bugs and/or Errors:
    None.
    
"""

from sys import exit
import random

# Defining the beginning of all the stages
def start():
    
    # Introduction to the user
    print(f"""
Hello and welcome to the Living Single inspired Game!
There are a total of five scenarios and the goal is to 
experience all the scenarios to see the storyline uncovered.
          
You're now in scenario_1 - On Set!\n
""")
    
    input(prompt = "<Press enter to continue>\n")
    scenario_1()
        
# Calling the first function in our map to start the game/story line
def scenario_1():
    
    print("Before we begin, choose your characters names.")
    
    #Defining three variables for the characters
    username_girl = input(prompt = "> Girl's name: ").capitalize()
    username_boy1 = input(prompt = "> First boy's name: ").capitalize()
    username_boy2 = input(prompt = "> Second boy's name: ").capitalize()
    
        
    # Continuing scenario_1
    print("Great, now the game begins!")

    input(prompt = "<Press enter to continue>\n")

    # This moves us into stage 1

    # Introduction of the story including the characters chosen above
    print(f"""
Once upon a time, the skies were blue and the sun was up, 
so it was time for {username_boy1}, the best weatherman in 
New York City, to deliver the forecasted news on Fox News.

""")
    
    # Engaging the user so that the story line does not become too long
    input(prompt = "<Press any key to continue>\n")
    
    print(f"""

Just 5 min before he was about to start, this new lady, 
with a beautiful black tight fitted dress and a long red coat, 
{username_girl}, brought {username_boy1} a cup of coffee.

{username_boy1} is mezmerized by {username_girl}'s beauty
and asks her on a date.
""")

    while True:

            # Creating a choice for the user to interact in the story
        print(f"""
What would you like to respond to {username_boy1}?

1) Yes, I would love to go on a date with you!
2) No thank you, I am not interested.
""")

        choice = input(prompt = ">  \n")

        # Choosing yes leads to the next stage (scenario_2)
        if "1" in choice or "yes" in choice or "date" in choice:
            print(f"""
\nYou both exchange numbers and plan a date for the evening! 
{username_girl} is happy and leaves the set.\n
""")
            input(prompt = "<Press enter to continue>\n")

            print(f"""
{username_boy1} starts the weather forecast and ends Fox News with the statement 
"Sorry ladies, Im taken tonight. 
This is {username_boy1}, Im a quiet storm keeping you warm." 

And we’re clear, to commercial!
""")
            
            # Engaging the user so that the story line does not become too long
            input(prompt = "<Press any key to continue>\n")

            ############################################################# 
            #Entering the next stage and using the same characters 
            #in the next stage
            #############################################################
            scenario_2(username_boy1 = username_boy1, 
                       username_boy2 = username_boy2, 
                       username_girl = username_girl)
            break

        # Choosing "No" ends the game
        elif "2" in choice or "no" in choice or "not interested" in choice:
            print(f"""\n{username_boy1} gets upset and {username_girl} 
            leaves the set.\n""")

            # This moves user into fail
            fail()
            break

        else:
            print("Invalid command. Please try again.")
            input(prompt = "<Press enter to continue>\n")
            continue

            
# Creating a fail function for user to end game upon specific choices below
def fail():
            print(f"""
After leaving the set she continued to stay single. 
Years went by and she understood that she had to be more open-minded
and give guys a chance to date her.

The End.
""")
        
            input(prompt = '<Press any key to exit>\n')
            
            
# Moving to stage 2
def scenario_2(username_boy1, username_boy2, username_girl):
    
    print("You're now in scenario_2 - The Elevator!\n")
    
    input(prompt = '<Press any key to continue>\n')
    
    # Continuing the story line where the third character is introduced
    print(f"""
While {username_boy1} delivers the weather forecast, 
{username_girl} goes out and bumps into a stranger. 

She loses her purse on the ground. 

An handsome 50-years old man walks out from the elevator and says 
”Allow me to help”. 
""")
    
    # Engaging the user so that the story line does not become too long
    input(prompt = "<Press any key to continue>\n")
    
    print(f"""

She finds out his name is {username_boy2}.
He is the architect of Empire State Building. 

Some flirting goes on, and {username_boy2} asks {username_girl} on a date.
""")
    
    while True:
        # Creating a choice for the user to interact in the story
        print(f"""
What would you like to respond to {username_boy2}?
1) Yes, I would love to go on a date with you!
2) No, thank you.
""")

        choice = input(">  \n")

        # Choosing yes leads to the next stage (scenario_3)
        if "1" in choice or "yes" in choice or "date" in choice:
            print(f"""\n
You both plan a date for tomorrow! 
“What should I wear?” says {username_girl}.

{username_boy2} replies: 
“Just go with anything that goes with that lovely smile!”

You smile and walk out the building.
\n""")

            input(prompt = "<Press enter to continue>\n")

            # Entering the next stage and using the same characters there
            scenario_3(username_boy1 = username_boy1, 
                       username_boy2 = username_boy2, 
                       username_girl = username_girl)
            break

        # Choosing "No" ends the game
        elif "2" in choice or "no" in choice or "thank you" in choice:
            print(f"""\n
{username_boy2} respectfully says good bye.
{username_girl} was happy with {username_boy1}. Or so she thought...

The End.
\n""")
            fail()
            break

        else:
            print("Invalid command. Please try again.")
            input(prompt = "<Press enter to continue>\n")
            
            # Brings us back to the beginning of scenario_2 to try again
            continue 

# Moving to stage 3
def scenario_3(username_boy1, username_boy2, username_girl):
    
    print(f"""
You're now in scenario_3 - Date #1 with {username_boy1}!
\n""")
    
    input(prompt = "<Press any key to continue>\n")
    
    # Continuing the story line on a date with username_boy1
    print(f"""
Evening approaches and {username_boy1} meets {username_girl} at Plaza Hotel. 
A waiter seats them down and hands them the menu.
They have a lot of options to choose from.
\n""")
    
    input(prompt = "<Press enter to continue>\n")
    
    print(f"""
{username_boy1}: I love this restaurant so much. 
There is one particular dish I love so much. Wanna guess which one?""")
    
    # declaring object
    menu_options = ['Sushi', 'Pasta', 'Taco', 'Pizza', 
                    'Hamburger', 'Wings']
    
    # Giving username_girl three guesses for the favorite dish of username_boy1
    guesses = 3
    
    # Choosing the favorite meal of boy 1 by randomness
    menu_choice = random.choice(menu_options)
    
    # Looping until a user is out of guesses
    while guesses > 0:
        
        print(f"""
These are the menu options:
1) Sushi
2) Pasta
3) Taco
4) Pizza
5) Hamburger
6) Wings
""")
        # Telling user to guess the right menu dish
        menu_guess = input(prompt = "<Guess the menu choice>\n").capitalize()
        
        if menu_guess in ['1', '1.', '1)', 'Sushi', 'S']: 
            menu_guess = 'Sushi'
            
        elif menu_guess in ['2', '2.', '2)', 'Pasta', 'P']: 
            menu_guess = 'Pasta'
            
        elif menu_guess in ['3', '3.', '3)', 'Taco', 'T']: 
            menu_guess = 'Taco'
        
        elif menu_guess in ['4', '4.', '4)', 'Pizza', 'P']: 
            menu_guess = 'Pizza'
        
        elif menu_guess in ['5', '5.', '5)', 'Hamburger', 'H']: 
            menu_guess = 'Hamburger'
            
        elif menu_guess in ['6', '6.', '6)', 'Wings', 'W']: 
            menu_guess = 'Wings'
        
        # Conditional statement based on user input
        if menu_guess == menu_choice:
            print(f"""
{username_boy1}: Wow, {menu_choice.lower()} is exactly my favorite dish! 
You got it right :D
""")
            break

        elif menu_guess != menu_choice: 
            if guesses == 3:
                print(f"""
{username_boy1}: Naah, that's not really my favorite dish babe. Try again. 
You have {guesses -1} guesses remaining.
""")
                
            elif guesses == 2:
                print(f"""
{username_boy1}: Not quite that either honey. 
You have {guesses -1} guess remaining.

Try again, you got this :D
""")
            
            elif guesses == 1:
                print(f"""
{username_boy1}: Haha, that's also not right. 

You are out of guesses my dear. My favorite dish is actually
{menu_choice.lower()}. But don't worry, let's order now.
""")

        # Setting a different error message for inner and outer loops
        else:
            print("\nPlease enter the dish correctly.")
            continue
            
        guesses -= 1

    # Ending scenario_3 on a good note
    print(f"""
{username_boy1} and {username_girl} enjoy the rest of the 
evening and date #1 is a success.
""")
    
    input(prompt = "<Press any key to continue>\n")
    
    # Bringing the characters to the next stage
    scenario_4(username_boy1 = username_boy1, username_boy2 = username_boy2, 
               username_girl = username_girl)

# Moving into scenario 4
def scenario_4(username_boy1, username_boy2, username_girl):
    
    print(f"""
You're now in scenario_4 - Date #2 with {username_boy2}!
\n""")
    
    input(prompt = "<Press any key to continue>\n")
    
    # Continuing the story line
    print(f"""
The next day arrives and {username_boy2} calls {username_girl}. 

They flirt and have a good conversation, but {username_boy2} wants
some suggestions on where they should go for their date.

He secretly already knows where he wants to take {username_girl} to, 
but wants to know her thoughts.

""")
    input(prompt = "<Press enter to continue>\n")
    
    # Declaring object for the different date options for the couple
    date_options = ['Restaurant', 'Cinema', 
                    'Painting class', 'Museum']
    
    # Giving username_girl two guesses for the date place usernameboy_2 wants
    guesses = 2
    
    # Choosing the right place for the date by randomness
    date_choice = random.choice(date_options)
    
    # Looping until the user is out of guesses
    while guesses > 0:
        
        print(f"""
These are the date options:
1) Restaurant
2) Cinema
3) Painting class
4) Museum
""")
        # Telling user to guess the right place for the date
        date_guess = input(prompt = "<Enter a date choice>\n").capitalize()
        
        if date_guess in ['1', '1.', '1)', 'Restaurant', 'R']:
            date_guess = 'Restaurant'
            
        elif date_guess in ['2', '2.', '2)', 'Cinema', 'C']:
            date_guess = 'Cinema'

        elif date_guess in ['3', '3.', '3)', 'Painting class', 'P']:
            date_guess = 'Painting class'

        elif date_guess in ['4', '4.', '4)', 'Museum', 'M']:
            date_guess = 'Museum'
            
        # Conditional statement based on user input
        if date_guess == date_choice:
            print(f"""
{username_boy2}: Wow, that's exactly what I was thinking! 
The {date_choice.lower()} is the best place I know of. How did you know? :D
""")
            break

        elif date_guess != date_options:
            if guesses == 2: 
                print(f"""
{username_boy2}: Hmm. What about the other options babe? 
You have {guesses -1} options left. Try again.
""") 
                
            elif guesses == 1:
                print(f"""
{username_boy2}: Hehe, I was actually thinking we could go to the {date_choice.lower()}. 
I have a surprise for you when we get there :)

But thanks for the feedback!
""")

        # Setting a different error message for inner and outer loops
        else:
            print(f"""\n
Please choose a date option again.""")
            continue
            
        guesses -= 1 
    
    # Continuing the storyline of date #2 with userboy_2
    print(f"""
{username_boy2} and {username_girl} have a great 
time at the {date_choice} and the evening is going along great.

{username_boy2}: That reminds me, I am having my birthday party 
next Friday since I turn 51 years hahah.

I would love for you to come and also meet my son. 

How does that sounds?
""")
    
    input(prompt = "<Press any key to continue>\n")
    
    print(f"""

{username_girl}: Oh, yes I would love to join!

{username_boy2} is happy. Date #2 has gone successful.
""")
    
    input(prompt = "<Press any key to continue>\n")
    
    # Continue to bring the characters to the final stage
    scenario_5(username_boy1 = username_boy1, username_boy2 = username_boy2, 
               username_girl = username_girl)
        
def scenario_5(username_boy1, username_boy2, username_girl):
    
    # Introducing the stage
    print(f"""
You're now in scenario_5, the final stage - The Birthday!
\n""")
    
    input(prompt = "<Press any key to continue>\n")
    
    # The Birthday reveal brings the storyline to a close
    # Displaying the interaction between the three characters
    print(f"""
Next Friday arrives and {username_girl} is ready 
to celebrate {username_boy2}'s birthday.

It is 7 pm and {username_boy2} picks {username_girl} up at 95th Street.
He pulls up to a grand mansion in Manhattan.

""")
    
    # Engaging the user so that the story line does not become too long
    input(prompt = "<Press any key to continue>\n")
    
    # Continuing the story line
    print(f"""

Once they get to the door, {username_girl} sees {username_boy1}.
She shouts: {username_boy1}! What are you doing here?

{username_boy1} says in shock: I am here to celebrate my father's birthday.
How do you know my dad?

{username_boy2}: Baby, how do you know my son {username_boy1}?

{username_girl}: YOUR DAD?! YOUR SON?!

""")
    
    # Engaging the user so that the story line does not become too long
    input(prompt = "<Press any key to continue>\n")
    
    # Continuing the story line
    print(f"""

After further discussion you find out you have been 
dating two people who happen to be related.

In fact you have been dating the father and son all along.

The men ask you to make a choice who you want to continue dating.
""")
    
    # Engaging the user so that the story line does not become too long
    input(prompt = "<Press any key to continue>\n")
    
    while True:
        
        # Creating a choice for user input
        print(f"""
Who do you choose to date?
1) {username_boy1}
2) {username_boy2}
""")
    
        choice_1 = input(prompt = ">  \n")

        # Conditional statement based on user input
        # Choosing the first option leads to a sad story & ends the game
        if "1" in choice_1 or f"{username_boy1}" in choice_1:
            print(f"""\n
After a couple of years of dating the weather forecast man {username_boy1}, 
you realize you made the wrong decision. 

You continue to look for your prince charming.
\n""")
            input(prompt = "<Press enter to continue>\n")
            exit(0)
            
        # Choosing the second option leads to a happy ending & ends the game
        elif "2" in choice_1 or f"{username_boy2}" in choice_1:
            print(f"""\n
This is your prince charming!
Even though {username_boy2} is old, he has a lot of
life experience and treats you like a princess.

You both live happily ever after in his grand mansion.
\n""")
            input(prompt = "<Press enter to continue>\n")
            exit(0)

        # In case user inputs something wrong, user will get the options again
        else:
            print("Invalid command. Please try again.")
            input("<Press enter to continue>\n")
            scenario_5() 
            
# Ends the whole game as we started the game with defining start
start()
```
